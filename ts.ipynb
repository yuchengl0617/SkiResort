```python
# ===================== Config =====================
FILE_PATH = "allianz.xlsx"     # adjust if needed
SHEET = "Visitation Data"

# Backtest & tuning controls
INTERVAL_WIDTH = 0.95
N_CV_YEARS = 3                 # use last K winters as test folds (if available)
MIN_TRAIN_POINTS = 20          # min training weekly points before a test winter
MIN_TEST_POINTS  = 10          # min weekly points in a test winter

# Hyperparameter grid (keep small first; expand if needed)
GRID_SEAS_PRIOR = [5.0, 10.0, 15.0, 20.0]
GRID_CP_PRIOR   = [0.05, 0.10, 0.20]
GRID_FOURIER    = [10, 15, 20]   # yearly fourier_order

# Output dirs
SAVE_DIR_CSV   = "visitor predictions"
SAVE_DIR_PLOTS = "visitor plots"
SAVE_DIR_TUNE  = "visitor predictions"  # put tuning CSVs with other CSVs

SHOW_BACKTEST_PLOTS = True  # show fold plots
SHOW_FINAL_PLOTS    = True  # show final plots

# ===================== Helpers =====================
def first_monday_on_or_after_june1(year: int) -> pd.Timestamp:
    d = datetime(year, 6, 1)
    offset = (0 - d.weekday()) % 7  # Monday=0
    return pd.Timestamp(d + timedelta(days=offset))

def make_season_dates(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    out["season_start"] = out["Year"].apply(first_monday_on_or_after_june1)
    out["ds"] = out["season_start"] + pd.to_timedelta((out["Week"] - 1) * 7, unit="D")
    return out

def build_prophet(seas_prior: float, cp_prior: float, fourier: int) -> Prophet:
    m = Prophet(
        yearly_seasonality=False,           # we will add yearly with chosen fourier
        weekly_seasonality=False,
        daily_seasonality=False,
        interval_width=INTERVAL_WIDTH,
        seasonality_prior_scale=seas_prior,
        changepoint_prior_scale=cp_prior,
    )
    m.add_seasonality(name="yearly", period=365.25, fourier_order=fourier)
    return m

def mae(y_true, y_pred) -> float:
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return float(np.mean(np.abs(y_true - y_pred)))

def rmse(y_true, y_pred) -> float:
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return float(np.sqrt(np.mean((y_true - y_pred) ** 2)))

def mape(y_true, y_pred) -> float:
    yt = np.asarray(y_true, dtype=float)
    yp = np.asarray(y_pred, dtype=float)
    mask = yt != 0
    if mask.sum() == 0:
        return np.nan
    return float(np.mean(np.abs((yt[mask] - yp[mask]) / yt[mask])) * 100.0)

def season_window(year: int, season_weeks: int):
    start = first_monday_on_or_after_june1(year)
    end   = start + pd.to_timedelta((season_weeks - 1) * 7, unit="D")
    return start, end

def ensure_dirs():
    os.makedirs(SAVE_DIR_CSV, exist_ok=True)
    os.makedirs(SAVE_DIR_PLOTS, exist_ok=True)
    os.makedirs(SAVE_DIR_TUNE, exist_ok=True)

# ===================== Load & Prepare =====================
vis = pd.read_excel(FILE_PATH, sheet_name=SHEET)

# Identify resort columns
resorts = [c for c in vis.columns if c not in ["Year", "Week"]]
vis[resorts] = vis[resorts].apply(pd.to_numeric, errors="coerce")

# Build season calendar dates
vis = make_season_dates(vis)

# Season length (weeks per winter, should be ~15)
season_weeks = int(vis["Week"].max())

last_year = int(vis["Year"].max())
next_year = last_year + 1
next_start, next_end = season_window(next_year, season_weeks)

ensure_dirs()

# ===================== Tuning + Forecast per resort =====================
best_rows = []     # best params per resort
all_tuning_rows = []  # detailed grid results per resort (averaged across folds)

for resort in resorts:
    # Build series for this resort
    df_r = (
        vis[["Year", "Week", "ds", resort]]
        .rename(columns={resort: "y"})
        .dropna()
        .sort_values("ds")
        .groupby("ds", as_index=False)
        .agg({"y": "sum"})
    )

    if len(df_r) < MIN_TRAIN_POINTS + MIN_TEST_POINTS:
        print(f"Skipping {resort}: not enough data points ({len(df_r)}).")
        continue

    # Determine test years: last K winters that have >= MIN_TEST_POINTS weeks
    counts_by_year = (
        df_r.assign(year=df_r["ds"].dt.year)
            .groupby("year")["ds"].count()
            .sort_index()
    )
    eligible_years = counts_by_year[counts_by_year >= MIN_TEST_POINTS].index.tolist()
    if not eligible_years:
        print(f"Skipping tuning for {resort}: no winter meets MIN_TEST_POINTS.")
        continue
    test_years = eligible_years[-N_CV_YEARS:]  # last K years

    # ---------- Grid Search ----------
    grid_results = []
    for seas_prior in GRID_SEAS_PRIOR:
        for cp_prior in GRID_CP_PRIOR:
            for fourier in GRID_FOURIER:
                fold_metrics = []
                for ty in test_years:
                    test_start, test_end = season_window(int(ty), season_weeks)
                    train_bt = df_r[df_r["ds"] < test_start].copy()
                    test_bt  = df_r[(df_r["ds"] >= test_start) & (df_r["ds"] <= test_end)].copy()

                    if len(train_bt) < MIN_TRAIN_POINTS or len(test_bt) < 2:
                        # skip this fold if not enough data for the split
                        continue

                    m_bt = build_prophet(seas_prior, cp_prior, fourier)
                    m_bt.fit(train_bt)

                    # Forecast through the end of test season (anchor Monday)
                    periods_needed_bt = max(1, ((test_end - train_bt["ds"].max()).days // 7) + 8)
                    future_bt = m_bt.make_future_dataframe(periods=periods_needed_bt, freq="W-MON")
                    fc_bt = m_bt.predict(future_bt)

                    # Normalize ds to Monday week starts for safe merge
                    test_bt_norm = test_bt.copy()
                    test_bt_norm["ds"] = test_bt_norm["ds"].dt.to_period("W-MON").dt.start_time
                    fc_slice = fc_bt[(fc_bt["ds"] >= test_start) & (fc_bt["ds"] <= test_end)].copy()
                    fc_slice["ds"] = fc_slice["ds"].dt.to_period("W-MON").dt.start_time

                    merged = pd.merge(test_bt_norm, fc_slice[["ds", "yhat"]], on="ds", how="inner")
                    if len(merged) >= 2:
                        _mae = mae(merged["y"], merged["yhat"])
                        _rmse = rmse(merged["y"], merged["yhat"])
                        _mape = mape(merged["y"], merged["yhat"])
                        fold_metrics.append((_mae, _rmse, _mape))

                        # Optional: visualize folds
                        if SHOW_BACKTEST_PLOTS and len(test_years) == 1:
                            fig_bt, ax = plt.subplots(figsize=(10,5))
                            ax.plot(train_bt["ds"], train_bt["y"], ".", markersize=3, label="Train")
                            ax.plot(merged["ds"], merged["y"], "o", markersize=4, label=f"Actual {ty}")
                            ax.plot(merged["ds"], merged["yhat"], "-", label=f"Pred {ty} ({seas_prior},{cp_prior},{fourier})")
                            ax.axvspan(test_start, test_end, alpha=0.1, label="Test window")
                            ax.set_title(f"{resort} — Fold {ty}")
                            ax.legend()
                            plt.show()
                            plt.close(fig_bt)

                if fold_metrics:
                    arr = np.array(fold_metrics)
                    mean_mae, mean_rmse, mean_mape = arr.mean(axis=0)
                    grid_results.append({
                        "resort": resort,
                        "seas_prior": seas_prior,
                        "cp_prior": cp_prior,
                        "fourier": fourier,
                        "folds": len(fold_metrics),
                        "MAE": float(mean_mae),
                        "RMSE": float(mean_rmse),
                        "MAPE_%": float(mean_mape),
                        "test_years": ",".join(map(str, test_years)),
                    })

    if not grid_results:
        print(f"{resort}: No valid folds produced for tuning — using default params.")
        best = {"seas_prior": 15.0, "cp_prior": 0.10, "fourier": 15, "RMSE": np.nan, "MAE": np.nan, "MAPE_%": np.nan}
    else:
        grid_df = pd.DataFrame(grid_results).sort_values(["RMSE", "MAE"]).reset_index(drop=True)
        # Save per-resort tuning table
        out_tune = os.path.join(SAVE_DIR_TUNE, f"{resort.replace(' ', '_')}_tuning_results.csv")
        grid_df.to_csv(out_tune, index=False)
        print(f"Saved tuning results for {resort} → {out_tune}")
        all_tuning_rows.extend(grid_df.to_dict(orient="records"))

        # Best row
        best_row = grid_df.iloc[0]
        best = {
            "seas_prior": float(best_row["seas_prior"]),
            "cp_prior": float(best_row["cp_prior"]),
            "fourier": int(best_row["fourier"]),
            "RMSE": float(best_row["RMSE"]),
            "MAE": float(best_row["MAE"]),
            "MAPE_%": float(best_row["MAPE_%"]),
            "folds": int(best_row["folds"]),
            "test_years": best_row["test_years"],
        }

    # Keep best summary
    best_rows.append({"resort": resort, **best})

    # ---------- Final model on all history with best params ----------
    m = build_prophet(best["seas_prior"], best["cp_prior"], best["fourier"])
    m.fit(df_r)

    periods_needed = ((next_end - df_r["ds"].max()).days // 7) + 8
    periods_needed = max(periods_needed, 52)
    future = m.make_future_dataframe(periods=periods_needed, freq="W-MON")
    forecast = m.predict(future)

    # Slice next winter
    mask = (forecast["ds"] >= next_start) & (forecast["ds"] <= next_end)
    subset = forecast.loc[mask, ["ds", "yhat", "yhat_lower", "yhat_upper"]].reset_index(drop=True)

    # Save CSV forecast
    out_csv = os.path.join(SAVE_DIR_CSV, f"{resort.replace(' ', '_')}_forecast_{next_year}.csv")
    subset.to_csv(out_csv, index=False)
    print(f"Saved forecast CSV for {resort} → {out_csv}")

    # Plot 1: Full forecast
    fig1 = m.plot(forecast)
    plt.title(f"{resort} — Full Forecast (best: seas={best['seas_prior']}, cp={best['cp_prior']}, F={best['fourier']})")
    plt.xlabel("Date"); plt.ylabel("Visitor days")
    out_plot1 = os.path.join(SAVE_DIR_PLOTS, f"{resort.replace(' ', '_')}_full_forecast_{next_year}.png")
    fig1.savefig(out_plot1, dpi=300, bbox_inches="tight")
    if SHOW_FINAL_PLOTS:
        plt.show()
    plt.close(fig1)

    # Plot 2: Zoomed next season
    fig2, ax = plt.subplots(figsize=(10,5))
    ax.plot(df_r["ds"], df_r["y"], ".", label="Historical", markersize=3)
    ax.plot(subset["ds"], subset["yhat"], label="Forecast")
    ax.fill_between(subset["ds"], subset["yhat_lower"], subset["yhat_upper"], alpha=0.3, label="Uncertainty")
    ax.set_title(f"{resort} — {next_year} season (best: seas={best['seas_prior']}, cp={best['cp_prior']}, F={best['fourier']})")
    ax.set_xlabel("Date"); ax.set_ylabel("Visitor days"); ax.legend()
    out_plot2 = os.path.join(SAVE_DIR_PLOTS, f"{resort.replace(' ', '_')}_zoomed_{next_year}.png")
    fig2.savefig(out_plot2, dpi=300, bbox_inches="tight")
    if SHOW_FINAL_PLOTS:
        plt.show()
    plt.close(fig2)

# ===================== Save overall summaries =====================
if best_rows:
    best_df = pd.DataFrame(best_rows).sort_values(["RMSE", "MAE"], na_position="last")
    best_path = os.path.join(SAVE_DIR_TUNE, "best_params_by_resort.csv")
    best_df.to_csv(best_path, index=False)
    print(f"\nSaved best params per resort → {best_path}")
    print(best_df)

if all_tuning_rows:
    all_df = pd.DataFrame(all_tuning_rows)
    all_path = os.path.join(SAVE_DIR_TUNE, "all_tuning_results.csv")
    all_df.to_csv(all_path, index=False)
    print(f"Saved all tuning results (combined) → {all_path}")
```
